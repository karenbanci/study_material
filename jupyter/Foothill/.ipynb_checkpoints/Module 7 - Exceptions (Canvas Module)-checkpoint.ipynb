### What happens in the following program if the user enter an alpha string instead of a number?

```python
import random 
secret_number = random.randint(1,5)
while True:
    user_guess = input("Pick an integer: ")
    
    if int(user_guess) == secret_number:
        print("Thanks, you got it!\n\n")
        break
    else:
        print("Nope, please guess again.")
        continue
```

### In the program below,   we have added a try-except block
- if the ValueError is raised by the int( ) function, then execution jumps to the except block.  
 - Note:  The ValueError is generated by Python
- If a ValueError is NOT raised (converting from string to int was successful), then the except block is skipped. 

```python
import random
low  = 1
high = 5
secret_number = random.randint(low,high)
while True:
    user_guess = input("Pick an integer: ")
    if int(user_guess) < low or int(user_guess) > high:
        print(f"guess must be between {low} and {high}")
        continue
    try:
        if int(user_guess) == secret_number:
            print("Thanks, you got it!\n\n")
            break
        else:
            print("Nope, please guess again.")
            continue
    except ValueError:
         print("You must type an integer.")
```

Why does the built-in function for changing something to an int would give a ValueError with a string as opposed to a TypeError. 
- Surely, it's expecting a number, right?

No it isn't. The int() function takes either a number or a string. It can convert the string "21" to the integer 21, but it can't convert "dog" to a number.

## User triggered exceptions
- Remember an exceptions can be triggered when something "exceptional" occurs

### Consider the function,  check_data()
- it is ensuring that value passed to it is positive

```python
""" 
Sets the values of the instance variables height and width,
assuming they are both positive.
"""
def check_data(value):
    if value <= 0:
        raise ValueError()
```

```python
height = int(input("Please enter the height: "))
try:
    check_data(height)
except ValueError:
    print("the height must be greater than 0")
```
